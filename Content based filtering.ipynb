```python
music_df = music_df.drop(labels=['Title', 'Description'], axis = 1)
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS

def remove_stopwords(text):
    filtered = [w.lower() for w in text if w.lower() not in STOP_WORDS and w.lower() not in ['tseries', 'star studios']]
    return " ".join(filtered)
```

```python
music_df['Combined_Text'] = music_df.apply(lambda x: remove_stopwords(x['Combined_Text']), axis=1)
```

```python
from gensim.models import Word2Vec
from numpy import triu

sentences = music_df['Combined_Text'].tolist()
w2v_model = Word2Vec(sentences, vector_size=100, window=5, min_count=2, workers=4)
```

```python
def get_song_vector(tokens, model):
    # Filter words that are in the vocabulary
    valid_words = [w for w in tokens if w in model.wv]
    if not valid_words:
        return np.zeros(model.vector_size)
    # Average vectors
    return np.mean([model.wv[w] for w in valid_words], axis=0)

music_df['embedding'] = music_df['Combined_Text'].apply(lambda x: get_song_vector(x, w2v_model))
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')  # You can also try 'median' or 'constant'
standard_column = imputer.fit_transform(standard_column)
```

```python
import numpy as np

combined_vectors = np.hstack([
    np.stack(music_df['embedding'].values),  # shape: (n_samples, embedding_dim)
    standard_column              # shape: (n_samples, num_numerical_features)
])
```

```python
combined_vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(combined_vectors)
```

```python
def recommend(index, top_n=5):
    sim_scores = similarity_matrix[index]
    top_indices = sim_scores.argsort()[-top_n-1:-1][::-1]
    return music_df.iloc[top_indices]['Video_ID']
```

```python
def recommend_with_scores(index, top_n=5):
    sim_scores = similarity_matrix[index]
    top_indices = sim_scores.argsort()[-top_n-1:-1][::-1]

    # Get the recommended rows as a DataFrame
    recommended = music_df.iloc[top_indices][['Video_ID']].copy()
    recommended['Similarity (%)'] = sim_scores[top_indices] * 100  # convert to percentage

    return recommended

# Print recommendations with similarity percentage:
def percentage(index, top_n):
    video_id = music_df.iloc[index]["Video_ID"]
    recommendations = recommend_with_scores(index, top_n)
    
    print(f'For Video ID: {video_id}')
    print(f"Link: https://music.youtube.com/watch?v={video_id}\n")
    
    for _, row in recommendations.iterrows():
        print(f"Video ID: {row['Video_ID']}")
        print(f"Link: https://music.youtube.com/watch?v={row['Video_ID']}")
        print(f"Similarity: {row['Similarity (%)']:.2f}%\n")
```

```python
recommend_with_scores(10)
```

```python
def recommend_least_similar(index, bottom_n=5):
    sim_scores = similarity_matrix[index]
    
    # Exclude the song itself (its similarity is 1 or 100%)
    bottom_indices = sim_scores.argsort()
    bottom_indices = bottom_indices[sim_scores[bottom_indices] > 0][:bottom_n]

    recommended = music_df.iloc[bottom_indices][['Video_ID']].copy()
    recommended['Similarity (%)'] = sim_scores[bottom_indices] * 100
    return recommended

def least_percentage(index, bottom_n):
    video_id = music_df.iloc[index]["Video_ID"]
    recommendations = recommend_least_similar(index, bottom_n)

    print(f'Original Video ID: {video_id}')
    print(f"Link: https://music.youtube.com/watch?v={video_id}\n")

    for _, row in recommendations.iterrows():
        print(f"Least Similar Video ID: {row['Video_ID']}")
        print(f"Link: https://music.youtube.com/watch?v={row['Video_ID']}")
        print(f"Similarity: {row['Similarity (%)']:.2f}%\n")
```
